```python
import pandas
df_hanfu_500_info=pandas.read_json(
    r"D:\fluent_python\load_loler\rank_500_info_20180724.txt",
    lines=True,
    encoding="utf-8")
print(df_hanfu_500_info.head())
print(len(df_hanfu_500_info))
```

```python
#首先删除哪些champions集合为空list的数据
df_hanfu_500_info=df_hanfu_500_info.drop(
    df_hanfu_500_info.index[df_hanfu_500_info.champions.str.len()<3])
print(df_hanfu_500_info.count())
print(df_hanfu_500_info.head())
```

```python
#缺失的数据集
df_missed=df_hanfu_500_info[df_hanfu_500_info.champions.str.get(-1)==""]
#不缺失的数据集
df_completed=df_hanfu_500_info[df_hanfu_500_info.champions.str.get(-1)!=""]
#分别打印数据完整和不完整的数据：
print("completed:",len(df_completed))
print("missed:",len(df_missed))
#打印缺失数据中的不重复数据
print("unduplicating data in missed data：",len(df_missed.name.unique()))
#打印缺失数据中与完整数据重复的缺失数据
print("duplicated missing data in completed data：",
      len([x for x in df_missed.name if x in df_completed.name]))
```

```python
df_hanfu_500_info.drop_duplicates("name",inplace=True)
print(df_hanfu_500_info.head())
print(len(df_hanfu_500_info))
```

```python
df_hanfu_500_info.index=df_hanfu_500_info.name
for cols in df_hanfu_500_info.columns:
    for num,i in enumerate("123"):
        item=cols+"_"+i
        df_hanfu_500_info[item]=df_hanfu_500_info[cols].str.get(num)
df_hanfu_500_info=df_hanfu_500_info.drop(
    ["Win","champions","kda","lose","name","win_rate","name_1","name_2","name_3"],
    axis=1)
df_hanfu_500_info.head()
```

```python
#胜率跟使用的英雄有关系吗？
df_hanfu_500_info["win_rate_1(%)"]=df_hanfu_500_info.win_rate_1.str.extract("(\d+)")
df_hanfu_500_info["win_rate_2(%)"]=df_hanfu_500_info.win_rate_2.str.extract("(\d+)")
df_hanfu_500_info["win_rate_3(%)"]=df_hanfu_500_info.win_rate_3.str.extract("(\d+)")
df_hanfu_500_info=df_hanfu_500_info.drop(["win_rate_1","win_rate_2","win_rate_3"],axis=1)
df_hanfu_500_info.head()
```

```python
for col in df_hanfu_500_info.columns[:3]:
     df_hanfu_500_info[col]=df_hanfu_500_info[col].astype("float")
for col in df_hanfu_500_info.columns[6:]:
     df_hanfu_500_info[col]=df_hanfu_500_info[col].astype("float")
df_hanfu_500_info.head()
```

```python
print("----")
# print(type(df_hanfu_500_info["kda_1"].str))
print("----")
print(type(df_hanfu_500_info["kda_2"].str))
print("----")
print(type(df_hanfu_500_info["kda_3"].str))
```

```python
print(len(df_hanfu_500_info[df_hanfu_500_info["kda_2"].str.contains("[a-zA-Z]+")]))
print(len(df_hanfu_500_info[df_hanfu_500_info["kda_3"].str.contains("[a-zA-Z]+")]))
```

```python
champions_1_list_kda=df_hanfu_500_info.loc[:,["champions_1","kda_1"]]
champions_2_list_kda=df_hanfu_500_info.loc[:,["champions_2","kda_2"]]
champions_3_list_kda=df_hanfu_500_info.loc[:,["champions_3","kda_3"]]
print(champions_1_list_kda.head())
print(champions_2_list_kda.head())
print(champions_3_list_kda.head())
```

```python
champions_1_list_kda.columns=["champions","kda"]
champions_2_list_kda.columns=["champions","kda"]
champions_3_list_kda.columns=["champions","kda"]
Cham_list_kda=champions_1_list_kda.append(champions_3_list_kda).append(champions_3_list_kda)
Cham_list_kda.head()
```

```python
Cham_list_kda=Cham_list_kda.drop(
   Cham_list_kda.index[Cham_list_kda.kda.str.contains("[a-zA-Z]+")==True]
)
Cham_list_kda=Cham_list_kda.drop(
   Cham_list_kda.index[Cham_list_kda.champions.str.contains("[\w]+")==False]
)
Cham_list_kda.head()
```

```python
import numpy as np
Cham_list_kda["kda"]=Cham_list_kda["kda"].astype("float")
Cham_list_kda=Cham_list_kda.groupby("champions")\
    ["kda"].aggregate({"count":len,"mean":np.mean})
Cham_list_kda.head()
```

```python
df_hanfu_500_info["win_rate_1(%)"]=df_hanfu_500_info["win_rate_1"].str.extract("(\d+)")
df_hanfu_500_info["win_rate_2(%)"]=df_hanfu_500_info["win_rate_2"].str.extract("(\d+)")
df_hanfu_500_info["win_rate_3(%)"]=df_hanfu_500_info["win_rate_3"].str.extract("(\d+)")
df_hanfu_500_info=df_hanfu_500_info.drop(["win_rate_1","win_rate_2","win_rate_3"],axis=1)
for col in df_hanfu_500_info.columns[:3]:
     df_hanfu_500_info[col]=df_hanfu_500_info[col].astype("int")
for col in df_hanfu_500_info.columns[9:]:
     df_hanfu_500_info[col]=df_hanfu_500_info[col].astype("float")
df_hanfu_500_info["sum_1"]=df_hanfu_500_info["Win_1"]+df_hanfu_500_info["lose_1"]
df_hanfu_500_info["sum_2"]=df_hanfu_500_info["Win_2"]+df_hanfu_500_info["lose_2"]
df_hanfu_500_info["sum_3"]=df_hanfu_500_info["Win_3"]+df_hanfu_500_info["lose_3"]
df_hanfu_500_info.head()
```

```python
champions_1_list_win_rate=df_hanfu_500_info.loc[:,["champions_1","win_rate_1(%)"]]
champions_2_list_win_rate=df_hanfu_500_info.loc[:,["champions_2","win_rate_2(%)"]]
champions_3_list_win_rate=df_hanfu_500_info.loc[:,["champions_3","win_rate_3(%)"]]

champions_1_list_sum=df_hanfu_500_info.loc[:,["champions_1","sum_1"]]
champions_2_list_sum=df_hanfu_500_info.loc[:,["champions_2","sum_2"]]
champions_3_list_sum=df_hanfu_500_info.loc[:,["champions_3","sum_3"]]

champions_1_list_win_rate.columns=["champions","wr"]
champions_2_list_win_rate.columns=["champions","wr"]
champions_3_list_win_rate.columns=["champions","wr"]

champions_1_list_sum.columns=["champions","sum_n"]
champions_2_list_sum.columns=["champions","sum_n"]
champions_3_list_sum.columns=["champions","sum_n"]

Cham_list_WinRate=champions_1_list_win_rate.\
append(champions_2_list_win_rate).append(champions_3_list_win_rate)

Cham_list_sum=champions_1_list_sum.\
append(champions_2_list_sum).append(champions_3_list_sum)
```

```python
#删除champions为空的行
Cham_list_WinRate=Cham_list_WinRate.drop(
   Cham_list_WinRate.index[Cham_list_WinRate.champions.str.contains("[\w]+")==False])
Cham_list_sum=Cham_list_sum.drop(
   Cham_list_sum.index[Cham_list_sum.champions.str.contains("[\w]+")==False])
#分类统计
Cham_list_WinRate=Cham_list_WinRate.groupby("champions")\
["wr"].aggregate({"count":len,"mean":np.mean})
Cham_list_Sum=Cham_list_sum.groupby("champions")\
["sum_n"].aggregate({"count":len,"mean":np.mean})
```
